```python
# Machine learning technique doesn't seem to be viable solution for this problem.
# In this notebook we will analyse the data in traditional way (statistical methods)
# to seek out the correlation between the strategies, properties of the system and the score
```

```python
# import libraries
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#df = pd.read_csv('dataset0.csv')  #Original dataset (unbalanced)
df = pd.read_csv('subset_data.csv')   # sampled data from full dataset (balanced)
df_hist = df[df['best_strategy']=='hist']
df_flat = df[df['best_strategy']=='flat']
df_dist = df[df['best_strategy']=='dist']
df_freq = df[df['best_strategy']=='freq']
print("History \n",df_hist.describe(),"\n")
print("Flat \n",df_flat.describe(),"\n")
print("Distance \n",df_dist.describe(),"\n")
print("Frquency \n",df_freq.describe(),"\n")
```

```python
# we can observe that Frequency strategy has the lowest average mean, 
# additionally the mean tot_test has the largest mean (14641)
# second largest 
```

```python

# Perform Shapiro-Wilk test
statistic, p_value = stats.shapiro(df_freq['tot_tests'])

# Print the test statistic and p-value
print("Test Statistic:", statistic)
print("p-value:", p_value)
```

```python

# Plot histogram
plt.hist(df_freq['tot_tests'], bins='auto', edgecolor='black')

# Add labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of FREQUENCE startegy')

# Show the plot
plt.show()
```

```python
# Plot histogram
plt.hist(df_hist['tot_tests'], bins='auto', edgecolor='black')

# Add labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of HISTORY strategy')

# Show the plot
plt.show()
```

```python
# Plot histogram
plt.hist(df_flat['tot_tests'], bins='auto', edgecolor='black')

# Add labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of FLAT strategy')

# Show the plot
plt.show()
```

```python
# Plot histogram
plt.hist(df_dist['tot_tests'], bins='auto', edgecolor='black')

# Add labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of DISTANCE strategy')

# Show the plot
plt.show()
```
